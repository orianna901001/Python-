K-Nearest Neighbor(KNN) 是一種無須機率分配的假設下的演算法，跟距離預測值最近的 k 個數值，來估計預測值。
1. 計算距離
2. 尋找最近 k 組數值
3. 類別型態資料以多數決

常見距離計算定義
1. 歐基里德距離（Euclidean distance)
2. 曼哈頓距離（Manhattan distance)
